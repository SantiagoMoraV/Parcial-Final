**Primer Algoritmo**

El primer algoritmo que utilice es bsicamente un cierto nmero de clculos matemticos bsicos con figuras geomtricas.
Entre dichos clculos se encuentra el algoritmo para hallar el rea y el permetro de un tringulo, de un cuadrado y de un pentgono simple. Adicionalmente a esto hay un calculo para hallar el porcentaje de un nmero.

El rea de un tringulo es igual a base por altura dividido por 2.
La altura es la recta perpendicular trazada desde un vrtice al lado opuesto (o su prolongacin). para hacerlo en necesaria la siguiente ecuacin:

A=$\frac{b*h}{2}$

El rea de un cuadrado es igual al cuadrado de la longitud del lado.  Se refiere a la multiplicacin de uno de sus lados por el otro, o simplemente de un lado al cuadrado, se entiende mejor con la siguiente ecuacin:

A=$a^{2}$

Para el caso del permetro, para hallarlo se multiplica por 4 el rea: 

P=4*a

Para los casos del pentgono, seran los siguientes clculos: 
rea y permetro a partir del lado (L).
rea y permetro a partir de la apotema (ap)


A=$\frac{5*L*ap}{2}$

P=5*L

La manera clsica de calcular un porcentaje es multiplicar la cantidad inicial por el porcentaje que se quiere conocer y, despus, dividir ese resultado por 100. Es decir:

Prcn = $\frac{porcentaje*nmero}{100}$
