**Primer Algoritmo**

El primer algoritmo que utilice es básicamente un cierto número de cálculos matemáticos básicos con figuras geométricas.
Entre dichos cálculos se encuentra el algoritmo para hallar el área y el perímetro de un triángulo, de un cuadrado y de un pentágono simple. Adicionalmente a esto hay un calculo para hallar el porcentaje de un número.

El área de un triángulo es igual a base por altura dividido por 2.
La altura es la recta perpendicular trazada desde un vértice al lado opuesto (o su prolongación). para hacerlo en necesaria la siguiente ecuación:

A=$\frac{b*h}{2}$

El área de un cuadrado es igual al cuadrado de la longitud del lado.  Se refiere a la multiplicación de uno de sus lados por el otro, o simplemente de un lado al cuadrado, se entiende mejor con la siguiente ecuación:

A=$a^{2}$

Para el caso del perímetro, para hallarlo se multiplica por 4 el área: 

P=4*a

Para los casos del pentágono, serían los siguientes cálculos: 
Área y perímetro a partir del lado (L).
Área y perímetro a partir de la apotema (ap)


A=$\frac{5*L*ap}{2}$

P=5*L

La manera clásica de calcular un porcentaje es multiplicar la cantidad inicial por el porcentaje que se quiere conocer y, después, dividir ese resultado por 100. Es decir:

Prcn = $\frac{porcentaje*número}{100}$

Sucesión de Fibonacci

Es una secuencia infinita de números naturales cuyos dos primeros términos son 1 y 1 y tal que, cualquier otro término se obtiene sumando los dos inmediatamente anteriores. Por tanto, se cumple la relación de recurrencia siguiente:

r<sub>1</sub> = 1,
r<sub>2</sub> = 1...
r<sub>n</sub> = r<sub>n-1</sub> + r<sub>n-2</sub> (r>2)
